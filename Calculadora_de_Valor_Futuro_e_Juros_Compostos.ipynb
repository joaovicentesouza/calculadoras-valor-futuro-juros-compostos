```python
###############################  CALCULADORA DE VALOR FUTURO  ###########################################################

# SUBSTITUA OS VALORES EM PARÊNTESE PELOS INDICADOS AO LADO:

valor_inicial = (0)  # Substitua pelo valor inicial do investimento
aporte_mensal_para_valor_futuro = (1000)  # Substitua pelo valor do aporte mensal
taxa_juros_mensal_para_valor_futuro = (0.01)  # Substitua pela taxa de juros mensal (por exemplo, 1% ao mês)
valor_futuro_desejado = (100000)  # Substitua pelo valor futuro desejado


#########################################################################################################################
def calcular_tempo_para_valor_futuro(valor_inicial, aporte_mensal, taxa_juros, valor_futuro):
    tempo_meses = 0
    while valor_inicial < valor_futuro:
        valor_inicial = valor_inicial * (1 + taxa_juros) + aporte_mensal
        tempo_meses += 1
    return tempo_meses

tempo_necessario = calcular_tempo_para_valor_futuro(valor_inicial, aporte_mensal_para_valor_futuro, taxa_juros_mensal_para_valor_futuro, valor_futuro_desejado)

print(f'O tempo necessário para atingir o valor futuro de R${valor_futuro_desejado:.2f} é de {tempo_necessario} meses ou {tempo_necessario/12:.0f} anos e {tempo_necessario%12:.0f} meses.')
```

```python
###############################  CALCULADORA DE JUROS COMPOSTOS  ###########################################################

# SUBSTITUA OS VALORES EM PARÊNTESE PELOS INDICADOS AO LADO:
principal_inicial = (0)  # Substitua pelo valor inicial do investimento
taxa_juros_mensal = (0.01)  # Substitua pela taxa de juros mensal (por exemplo, 1% ao mês)
aporte_mensal = (1000)  # Substitua pelo valor do aporte mensal
periodo_meses = (70) + 1  # Substitua pelo número de meses que deseja calcular

def calcular_capital_final(principal, taxa_juros, aporte_mensal, periodo_meses):
    total_aportes = 0
    for _ in range(periodo_meses):
        principal = principal * (1 + taxa_juros) + aporte_mensal
        total_aportes += aporte_mensal

    total_juros = principal - (principal_inicial + total_aportes)
    return principal, total_aportes, total_juros

evolucao_capital, total_aportes, total_juros = calcular_capital_final(principal_inicial, taxa_juros_mensal, aporte_mensal, periodo_meses)

print(f'O capital final após {periodo_meses} meses é: R${evolucao_capital:.2f}')
print(f'O total de aportes realizados é: R${total_aportes:.2f}')
print(f'O total de juros gerados é: R${total_juros:.2f}\n\n')


# Criando gráfico
meses = list(range(1, periodo_meses + 1))

# Calculando juros e aportes mês a mês
evolucao_capital_por_mes = [calcular_capital_final(principal_inicial, taxa_juros_mensal, aporte_mensal, mes)[0] for mes in meses]
total_aportes_por_mes = [calcular_capital_final(principal_inicial, taxa_juros_mensal, aporte_mensal, mes)[1] for mes in meses]
total_juros_por_mes = [calcular_capital_final(principal_inicial, taxa_juros_mensal, aporte_mensal, mes)[2] for mes in meses]

# Plotando as linhas correspondentes
plt.plot(meses, evolucao_capital_por_mes, label='Capital Final')
plt.plot(meses, total_aportes_por_mes, label='Total de Aportes', linestyle='--')
plt.plot(meses, total_juros_por_mes, label='Total de Juros', linestyle='--')

plt.xlabel('Meses')
plt.ylabel('Valor')
plt.title('Evolução do Investimento')
plt.legend()
plt.show()
```
